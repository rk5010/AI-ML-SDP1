```python
# Using KNN Imputer to address the Driver_Age missing values

# KNNImputer(missing_values=np.nan, n_neighbors=5, weights='uniform', metric='nan_euclidean', 
# copy=True, add_indicator=False)

from sklearn.impute import KNNImputer

imputer_int = KNNImputer(missing_values=np.nan)

# Fill the missing values for 'Driver_Age'

claims_data_F1['Driver_Age'] = imputer_int.fit_transform(claims_data_F1[['Driver_Age']])

claims_data_F1['Driver_Age'] = claims_data_F1['Driver_Age'].astype(int)

claims_data_F1
```

```python
# Fill the missing values for 'Premium_AMT'

claims_data_F1['Premium_AMT'] = imputer_int.fit_transform(claims_data_F1[['Premium_AMT']])

claims_data_F1
```

```python
# Fill the missing values for 'Claim_AMT'

claims_data_F1['Claim_AMT'] = imputer_int.fit_transform(claims_data_F1[['Claim_AMT']])

claims_data_F1
```

```python
# Fill null values of Policy_Num with actual value '123456789'  

claims_data_F1 = claims_data_F1.fillna({'Policy_Num': 123456789})

claims_data_F1
```
